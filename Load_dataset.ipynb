```python
import pandas as pd
import numpy as np
import sqlite3

# 1. Create Sample Data
customers_data = {
    'customer_id': [1, 2, 3, 4, 5, 5], # Includes a duplicate
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Eve'],
    'email': ['alice@email.com', 'bob@email.com', None, 'david@email.com', 'eve@email.com', 'eve@email.com'],
    'signup_date': ['2023-01-15', '2023-02-20', '2023-03-05', '2023-04-10', '2023-05-12', '2023-05-12']
}

products_data = {
    'product_id': [101, 102, 103, 104],
    'product_name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'category': ['Electronics', 'Accessories', 'Accessories', 'Electronics'],
    'price': [1200.00, 25.50, 75.00, 300.00]
}

orders_data = {
    'order_id': [1001, 1002, 1003, 1004, 1005],
    'customer_id': [1, 2, 1, 4, 99], # 99 is an orphaned record
    'product_id': [101, 102, 104, 103, 101],
    'order_date': ['2023-06-01', '2023-06-02', '2023-06-02', '2023-06-05', '2023-06-06'],
    'quantity': [1, 2, 1, 1, None] # Includes a null value
}

# Load into Pandas DataFrames
df_customers = pd.DataFrame(customers_data)
df_products = pd.DataFrame(products_data)
df_orders = pd.DataFrame(orders_data)

# Load into SQLite for SQL demonstrations
conn = sqlite3.connect(':memory:')
df_customers.to_sql('customers', conn, index=False)
df_products.to_sql('products', conn, index=False)
df_orders.to_sql('orders', conn, index=False)
```
