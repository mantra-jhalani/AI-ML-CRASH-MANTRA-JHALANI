```python
import pandas as pd

# --- Setup: Creating a dummy dataset of app user activity ---
data = {
    'User_ID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Primary_Crop': ['Wheat', 'Rice', 'Wheat', 'Maize', 'Wheat', 'Rice', 'Sugarcane', 'Wheat'],
    'Questions_Asked': [5, 12, 3, 8, 2, 15, 6, 4],
    'Session_Duration_Mins': [10.5, 25.0, 5.2, 18.0, 4.5, 30.5, 12.0, 8.0]
}
df = pd.DataFrame(data)


# Task 7: Produce Quick Insights with describe() and value_counts()

print("=== Quick Data Insights ===\n")

# 1. Run describe() on numeric columns
print("--- 1. Numeric Summary (describe) ---")
numeric_summary = df[['Questions_Asked', 'Session_Duration_Mins']].describe()
print(numeric_summary)

print("\n> OBSERVATION:")
print("> The average (mean) session duration is roughly 14 minutes, but the maximum is 30.5 minutes, showing a wide range in how long users engage with the assistant.")
print("-" * 50 + "\n")


# 2. Run value_counts() on a categorical column
print("--- 2. Categorical Summary (value_counts) ---")
crop_counts = df['Primary_Crop'].value_counts()
print(crop_counts)

print("\n> OBSERVATION:")
print("> Wheat is the dominant focus for this user subset, accounting for exactly half (4 out of 8) of all primary crops logged.")
print("-" * 50)
```
