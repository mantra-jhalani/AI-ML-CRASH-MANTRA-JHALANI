```python
import math
from functools import total_ordering

# ==========================================
# EXPLORE: @total_ordering
# By adding this decorator, Python automatically generates __le__, __gt__, 
# and __ge__ based on our __eq__ and __lt__ methods, saving us writing 3 extra functions!
# ==========================================

@total_ordering
class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")
        
        # Simplify the fraction immediately upon creation
        g = math.gcd(numerator, denominator)
        self.numerator = numerator // g
        self.denominator = denominator // g
        
        # Ensure the denominator is always positive (e.g., 1/-2 becomes -1/2)
        if self.denominator < 0:
            self.numerator = -self.numerator
            self.denominator = -self.denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

    def __add__(self, other):
        """Adds two fractions: (a/b) + (c/d) = (ad + bc) / bd"""
        new_num = (self.numerator * other.denominator) + (other.numerator * self.denominator)
        new_den = self.denominator * other.denominator
        
        # We return a new Fraction object. It will automatically simplify in __init__.
        return Fraction(new_num, new_den)

    def __eq__(self, other):
        """Checks if two fractions are equal."""
        # Since they are always simplified in __init__, we just compare exact parts
        return self.numerator == other.numerator and self.denominator == other.denominator

    def __lt__(self, other):
        """Checks if this fraction is strictly less than another (a/b < c/d)."""
        # Cross-multiplication avoids floating-point inaccuracies
        return (self.numerator * other.denominator) < (other.numerator * self.denominator)


# Driver Code to Test the Module

if __name__ == "__main__":
    
    # Pair 1: Basic addition and simplification
    f1 = Fraction(1, 2)
    f2 = Fraction(1, 4)
    print(f"Pair 1 Add: {f1} + {f2} = {f1 + f2}")  # Expect 3/4
    
    # Pair 2: Testing Equality and Auto-Simplification
    f3 = Fraction(2, 3)
    f4 = Fraction(4, 6)
    print(f"Pair 2 Eq: {f3} == {f4} is {f3 == f4}") # Expect True
    
    # Pair 3: Testing Less Than
    f5 = Fraction(3, 4)
    f6 = Fraction(4, 5)
    print(f"Pair 3 Lt: {f5} < {f6} is {f5 < f6}")   # Expect True
    
    # Testing the @total_ordering generated method (__ge__)
    print(f"Explore Test (>=): {f6} >= {f5} is {f6 >= f5}") # Expect True
```
