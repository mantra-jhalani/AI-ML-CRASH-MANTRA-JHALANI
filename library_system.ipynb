```python
# Base Class

class LibraryItem:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    def describe(self):
        return f"'{self.title}' by {self.author} ({self.year})"


# Child Classes

class Book(LibraryItem):
    def __init__(self, title, author, year, pages):
        # Call the parent class's __init__ method to handle common attributes
        super().__init__(title, author, year)
        self.pages = pages

    # Override the parent's describe method
    def describe(self):
        # We can use super().describe() to get the base string and add to it
        base_desc = super().describe()
        return f"{base_desc} - Paperback: {self.pages} pages"

class EBook(LibraryItem):
    def __init__(self, title, author, year, file_size_mb):
        # Call the parent class's __init__ method
        super().__init__(title, author, year)
        self.file_size_mb = file_size_mb

    # Override the parent's describe method
    def describe(self):
        base_desc = super().describe()
        return f"{base_desc} - Digital: {self.file_size_mb}MB"


# Driver Code (TASK)

# Create 2 Book objects
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", 1925, 180)
book2 = Book("1984", "George Orwell", 1949, 328)

# Create 2 EBook objects
ebook1 = EBook("Digital Minimalism", "Cal Newport", 2019, 2.5)
ebook2 = EBook("Atomic Habits", "James Clear", 2018, 3.1)

# Store all four in a single list
library_collection = [book1, book2, ebook1, ebook2]

print("--- Library Collection ---")
# Loop through and call describe() on each
for item in library_collection:
    print(item.describe())



# EXPLORE: isinstance() behavior

print("\n--- Explore Task ---")

is_item = isinstance(book1, LibraryItem)
print(f"Is book1 a LibraryItem? {is_item}")

"""
Explanation for EXPLORE:
The result is True. In Object-Oriented Programming, inheritance sets up an 
"is-a" relationship. Because the Book class inherits from LibraryItem, every 
Book object is also considered a LibraryItem. The isinstance() function checks 
not just the immediate class, but the entire inheritance chain.
"""
```
