```python
from typing import Optional


# Core Calculator Operations


def add(a: float, b: float) -> float:
    """
    Adds two numbers together.
    
    Args:
        a (float): The first number.
        b (float): The second number.
        
    Returns:
        float: The sum of a and b.
    """
    return a + b

def subtract(a: float, b: float) -> float:
    """
    Subtracts the second number from the first.
    
    Args:
        a (float): The number to be subtracted from.
        b (float): The number to subtract.
        
    Returns:
        float: The difference between a and b.
    """
    return a - b

def multiply(a: float, b: float) -> float:
    """
    Multiplies two numbers.
    
    Args:
        a (float): The first number.
        b (float): The second number.
        
    Returns:
        float: The product of a and b.
    """
    return a * b


# Modified Operations (EXPLORE TASK)


def divide(a: float, b: float) -> Optional[float]:
    """
    Divides the first number by the second. Returns None if dividing by zero.
    
    Args:
        a (float): The dividend.
        b (float): The divisor.
        
    Returns:
        Optional[float]: The quotient of a / b, or None if b is 0.
    """
    if b == 0:
        return None
    return a / b


# New Operations (TASK)


def power(base: float, exp: float) -> float:
    """
    Raises a base number to a specific exponent.
    
    Args:
        base (float): The base number.
        exp (float): The exponent to raise the base to.
        
    Returns:
        float: The result of base raised to the power of exp.
    """
    return base ** exp

def modulo(a: int, b: int) -> int:
    """
    Calculates the remainder of division between two integers.
    
    Args:
        a (int): The dividend.
        b (int): The divisor.
        
    Returns:
        int: The remainder of a / b.
        
    Raises:
        ValueError: If the divisor (b) is 0.
    """
    if b == 0:
        raise ValueError("Cannot perform modulo by zero.")
    return a % b


# Driver Code to Test the Module

if __name__ == "__main__":
    print(f"Addition (5.5 + 2.0): {add(5.5, 2.0)}")
    print(f"Power (2.0 ^ 3.0): {power(2.0, 3.0)}")
    
    # Testing the Optional return on division
    safe_div = divide(10.0, 0.0)
    print(f"Safe Division by Zero (10.0 / 0.0): {safe_div}")
    
    # Testing modulo zero handling
    try:
        mod_result = modulo(10, 0)
    except ValueError as e:
        print(f"Modulo Error Caught: {e}")
```
