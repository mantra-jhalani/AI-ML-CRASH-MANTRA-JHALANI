```python
class InternProject:
    """Represents an intern's assigned project in the workflow."""
    
    def __init__(self, intern_name: str, project_name: str, primary_tech: str):
        # Initializing the instance attributes
        self.intern_name = intern_name
        self.project_name = project_name
        self.primary_tech = primary_tech
        self.status = "In Development"
        
    def get_project_summary(self) -> str:
        """Returns a formatted string summarizing the project status."""
        # Using f-strings to format the output cleanly
        return (
            f"=== Project Status Report ===\n"
            f" Intern  : {self.intern_name}\n"
            f" Project : {self.project_name}\n"
            f" Stack   : {self.primary_tech}\n"
            f" Status  : {self.status}\n"
            f"============================="
        )

if __name__ == "__main__":
    # 1. Instantiate the class with relevant real-world data
    current_project = InternProject(
        intern_name="Mantra Jhalani",
        project_name="Krishi mitra",
        primary_tech="Python & Flutter"
    )

    # 2. Call the method and print the output
    print(current_project.get_project_summary())
```
