```python
import pandas as pd
import os

# --- Setup ---
# Creating a dummy CSV file representing agricultural schemes data.
csv_content = """Scheme_Name,Category,State,Active_Users,Funding_Tier
PM-Kisan,Financial,Rajasthan,120000,High
Kusum Yojana,Solar,Gujarat,45000,Medium
Fasal Bima,Insurance,Rajasthan,85000,High
Krishi Sinchai,Irrigation,Punjab,30000,Medium
Paramparagat Krishi,Organic,Rajasthan,15000,Low
"""
with open("schemes_data.csv", "w") as f:
    f.write(csv_content)


# Task 4: Select Specific Columns and Filter Rows from a DataFrame


def process_dataframe(filepath: str):
    # 1. Load the small CSV dataset into Pandas
    df = pd.read_csv(filepath)
    
    print("--- 1. Original Dataset ---")
    print(df)
    print("\n")

    # 2. Select two or three specific columns
    # We are isolating just the scheme name, the state, and the user count.
    selected_columns_df = df[['Scheme_Name', 'State', 'Active_Users']]
    
    # 3. Filter rows with simple conditions (Boolean Filtering)
    # Filter A: Keep only rows where the State is 'Rajasthan'
    # Filter B: Keep only rows where Active_Users is greater than 50,000
    condition_state = selected_columns_df['State'] == 'Rajasthan'
    condition_users = selected_columns_df['Active_Users'] > 50000
    
    # Applying both conditions using the bitwise AND operator (&)
    filtered_df = selected_columns_df[condition_state & condition_users]

    # 4. Print the filtered result and note what each filter is doing
    print("--- 2. Filtered Result ---")
    print("Filters Applied: State is 'Rajasthan' AND Active Users > 50,000")
    print("-" * 55)
    print(filtered_df)
    print("-" * 55)

if __name__ == "__main__":
    process_dataframe("schemes_data.csv")
    
    # --- Cleanup ---
    # Removing the temporary CSV file after execution
    if os.path.exists("schemes_data.csv"):
        os.remove("schemes_data.csv")
```
