```python
import pandas as pd
import numpy as np

# --- Setup: Creating a dataset with intentional missing values ---
data = {
    'User_ID': [101, 102, 103, 104, 105],
    'App_Version': ['v1.2', np.nan, 'v1.3', 'v1.2', np.nan],
    'Session_Time_Mins': [15.5, 22.0, np.nan, 12.5, 30.0],
    'Feedback_Score': [4.0, 5.0, 3.0, np.nan, 4.0]
}
df = pd.DataFrame(data)

print("=== 1. Original Dataset (Before) ===")
print(df)
print("\n" + "-"*40 + "\n")


# Task 6: Clean Missing Values and Inspect the Dataset

# 1. Run isnull().sum() to identify columns with missing values
print("=== 2. Missing Values Inspection ===")
missing_counts = df.isnull().sum()
print("Total missing values per column:")
print(missing_counts)
print("\n" + "-"*40 + "\n")

# 2. Use dropna() for one case
# Strategy: Drop rows where a critical piece of information ('App_Version') is missing.
# We use .copy() to avoid Pandas warnings when modifying the new DataFrame later.
df_cleaned = df.dropna(subset=['App_Version']).copy()

# 3. Use fillna() for another case
# Strategy A: Fill missing 'Session_Time_Mins' with the average (mean) time.
mean_time = df_cleaned['Session_Time_Mins'].mean()
df_cleaned['Session_Time_Mins'] = df_cleaned['Session_Time_Mins'].fillna(mean_time)

# Strategy B: Fill missing 'Feedback_Score' with a default value of 0.0.
df_cleaned['Feedback_Score'] = df_cleaned['Feedback_Score'].fillna(0.0)

# 4. Show the after result
print("=== 3. Cleaned Dataset (After dropna & fillna) ===")
print(df_cleaned)
```
