```python
import json

# EXPLORE TASK: dump() vs dumps()
# ==========================================
# json.dump() serializes a Python object and writes it directly to a file object.
# json.dumps() serializes a Python object and returns it as a formatted string in memory.

# ==========================================
# Core Functions
# ==========================================

def save_config(data: dict, filename: str):
    """Writes a dictionary to a JSON file."""
    # Using 'w' mode to write/overwrite the file
    with open(filename, 'w') as file:
        # indent=4 makes the JSON file human-readable
        json.dump(data, file, indent=4)

def load_config(filename: str) -> dict:
    """Reads a JSON file and returns a dictionary."""
    try:
        # Using 'r' mode to read the file
        with open(filename, 'r') as file:
            return json.load(file)
    except FileNotFoundError:
        print(f"Warning: {filename} not found. Returning empty config.")
        return {}

def update_config(filename: str, key: str, value):
    """Loads the config, updates a single key, and saves it back."""
    # 1. Load the existing data
    config_data = load_config(filename)
    
    # 2. Update the specific key
    config_data[key] = value
    
    # 3. Save the modified dictionary back to the file
    save_config(config_data, filename)


# Driver Code to Test the Module

if __name__ == "__main__":
    config_file = "app_config.json"
    
    # Initial configuration setup for model training
    initial_settings = {
        "model": "krishi_mitra_vision_v1",
        "learning_rate": 0.001,
        "epochs": 10
    }
    
    print("--- 1. Saving Initial Config ---")
    save_config(initial_settings, config_file)
    print(f"Saved initial configuration to {config_file}.")
    
    print("\n--- 2. Updating Config ---")
    update_config(config_file, "epochs", 20)
    print("Updated 'epochs' to 20.")
    
    print("\n--- 3. Verifying the Update ---")
    final_config = load_config(config_file)
    print("Current Configuration state:")
    
    # Using json.dumps here to print the dictionary nicely to the console
    print(json.dumps(final_config, indent=4))
```
