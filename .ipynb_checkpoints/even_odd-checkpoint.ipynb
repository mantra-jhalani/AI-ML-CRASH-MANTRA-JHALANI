```python
# EXPLORE: What does int() do when given a string like "abc"?
# Answer: It raises a "ValueError". The int() function expects a string that 
# contains only numbers (like "42"). When it sees letters, it doesn't know 
# how to convert them into a base-10 number, so it throws an error and stops.

def main():
    # Ask the user for input
    user_text = input("Please enter the number: ")

    try:
        # Try to convert the text into an integer
        number = int(user_text)

        # Check conditions using if / elif / else
        if number == 0:
            print("The number is zero.")
        elif number % 2 == 0:
            print("The number is even.")
        else:
            print("The number is odd.")

    except ValueError:
        # This catches the error if the user types something like "abc"
        print("That is not a valid number. Please run the script and try again.")
  
if __name__ == "__main__":
    main() 
   
```
