```python
def main():
    user_text = input("Please enter a number: ")

    # The 'try' and 'except' must align perfectly vertically
    try:
        number = int(user_text)
        
        if number == 0:
            print("The number is zero.")
        elif number % 2 == 0:
            print("The number is even.")
        else:
            print("The number is odd.")

    except ValueError:
        # Don't forget to indent the print statement inside the except block too!
        print("That is not a valid number. Please run the script and try again.")

        # This is the part that actually TELLS the code to run!
if __name__ == "__main__":
    main()
```
