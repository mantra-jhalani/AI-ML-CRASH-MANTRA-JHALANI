```python
from typing import Dict

# 1. Store learner profile using variables and a dictionary
student_name = "Mantra Jhalani"
student_email = "mantrajhalani@gmail.com"

academic_profile = {
    "program": "B.Tech Software Engineering",
    "semester": "IV",
    "primary_languages": "Python, C++, Flutter"
}

# 2. Use a function with type hints returning a formatted profile card
def generate_profile_card(name: str, email: str, details: Dict[str, str]) -> str:
    """Generates a 4-line student profile card."""
    
    # 3. Use f-strings for every line of output
    line_1 = f"====================================="
    line_2 = f" NAME   : {name} | {email}"
    line_3 = f" DEGREE : {details['program']} (Sem {details['semester']})"
    line_4 = f" SKILLS : {details['primary_languages']}"
    
    # Combine lines with newline characters
    return f"{line_1}\n{line_2}\n{line_3}\n{line_4}\n{line_1}"

# Execute the function and print the deliverable
if __name__ == "__main__":
    profile_summary = generate_profile_card(student_name, student_email, academic_profile)
    print(profile_summary)
```
