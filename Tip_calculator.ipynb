```python
# EXPLORE: What's the difference between return and print?
# print() just displays text on the screen for the user to read.
# return passes data back to the program so the computer can store or use it later.

def calculate_tip(bill , tip_percent):
    # Calculate the exact tip amount (converting the whole number percentage to a decimal)
    tip_amount = bill * (tip_percent / 100)

    # Calculate the final total
    total = bill + tip_amount

    # Pack the results into a dictionary and return it
    return{"tip": tip_amount, "total": total}

def main():
    # Test 1
    result1 = calculate_tip(50.00, 15)
    print(f"Test 1 - Bill $50, 15% Tip: {result1}")

    # Test 2
    result2 = calculate_tip(85.50, 20)
    print(f"Test 2 - Bill $85.50, 20% Tip: {result2}")

    # Test 3
    result3 = calculate_tip(10.00, 10)
    print(f"Test 3 - Bill $10, 10% Tip: {result3}")

if __name__ == "__main__":
    main()
```
