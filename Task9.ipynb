```python
import numpy as np

# =====================================================================
# Task 9: Use Masking, Broadcasting, and a Similarity Calculation
# =====================================================================

print("=== 1. Boolean Masking ===")
# Generate a standard array of integers
data_array = np.array([12, 45, 67, 23, 89, 34, 10, 55])
print(f"Original Array: {data_array}")

# Create a boolean mask to filter values greater than 40 without any loops
mask = data_array > 40
filtered_data = data_array[mask]

print(f"Boolean Mask (> 40): {mask}")
print(f"Filtered Array: {filtered_data}")
print("\n" + "-"*50 + "\n")


print("=== 2. Broadcasting ===")
# Apply a broadcasted operation to scale the array
# This multiplies every single element by 1.5 in one compiled C-level operation
scaled_array = data_array * 1.5

print(f"Broadcasted Array (Original * 1.5):\n{scaled_array}")
print("\n" + "-"*50 + "\n")


print("=== 3. Cosine Similarity Calculation ===")
def cosine_similarity(v1, v2):
    """Calculates the cosine similarity between two NumPy vectors."""
    # Implementing the standard formula using dot product and magnitudes (norms)
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    
    # Cosine similarity = (A . B) / (||A|| * ||B||)
    return dot_product / (norm_v1 * norm_v2)

# Test Pair 1: Highly similar vectors (Should be close to 1.0)
vector_a = np.array([3.0, 4.0, 5.0])
vector_b = np.array([3.1, 3.9, 5.2])
sim_1 = cosine_similarity(vector_a, vector_b)

# Test Pair 2: Orthogonal/Perpendicular vectors (Should be exactly 0.0)
vector_c = np.array([1.0, 0.0, 0.0])
vector_d = np.array([0.0, 1.0, 0.0])
sim_2 = cosine_similarity(vector_c, vector_d)

print("Test Pair 1 (Similar Vectors):")
print(f" Vector A: {vector_a} | Vector B: {vector_b}")
print(f" Cosine Similarity: {sim_1:.4f}\n")

print("Test Pair 2 (Orthogonal Vectors):")
print(f" Vector C: {vector_c} | Vector D: {vector_d}")
print(f" Cosine Similarity: {sim_2:.4f}")
print("\n=====================================================================")
```
