```python
# TASK: List Comprehension Drills

# 1. Extract integers divisible by 3 from a list of 20 integers
numbers = list(range(1, 21)) # Creates a list from 1 to 20
divisible_by_3 = [num for num in numbers if num % 3 == 0]
print(f"1. Divisible by 3: {divisible_by_3}")

# 2. Return words longer than 4 characters in title case
words_list = ["code", "python", "dev", "interface", "app", "algorithm"]
long_words = [word.title() for word in words_list if len(word) > 4]
print(f"2. Long Words (Title Case): {long_words}")

# 3. Convert all Celsius temperatures to Fahrenheit
celsius_temps = [0, 15, 25, 30, 100]
fahrenheit_temps = [(temp * 9/5) + 32 for temp in celsius_temps]
print(f"3. Fahrenheit Temps: {fahrenheit_temps}")

# 4. Flatten a nested list
nested = [[1, 2], [3, 4], [5, 6], [7, 8]]
flattened = [num for sublist in nested for num in sublist]
print(f"4. Flattened List: {flattened}")

print("\n" + "="*40 + "\n")


# EXPLORE: Dict and Set Comprehensions

# Dictionary Comprehension Example
# Creating a dictionary where the key is a number (1-5) and the value is its square
squares_dict = {x: x**2 for x in range(1, 6)}
print(f"Dict Comprehension (Squares): {squares_dict}")

# Set Comprehension Example
# Extracting unique vowels from a string (notice duplicate 'e' is removed)
sample_text = "software engineering"
unique_vowels = {char for char in sample_text if char in "aeiou"}
print(f"Set Comprehension (Unique Vowels): {unique_vowels}")
```
