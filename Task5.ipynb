```python
import pandas as pd

# --- Setup: Creating a small dataset with custom row labels ---
data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headset'],
    'Price': [1200, 25, 75, 300, 80],
    'Stock': [15, 100, 45, 20, 60],
    'Rating': [4.5, 4.2, 4.8, 4.6, 4.1]
}

# Setting explicit string labels for the index instead of default 0, 1, 2...
df = pd.DataFrame(data, index=['INV-01', 'INV-02', 'INV-03', 'INV-04', 'INV-05'])

print("=== Original Dataset ===")
print(df)
print("\n" + "="*40 + "\n")


# Task 5: Compare .loc and .iloc

# 1. Using .loc (Label-based selection)
# Selecting rows by their index name and columns by their column name.
loc_selection = df.loc[['INV-01', 'INV-03'], ['Product', 'Price']]

print("=== Selection using .loc ===")
print("Query: df.loc[['INV-01', 'INV-03'], ['Product', 'Price']]")
print("-" * 30)
print(loc_selection)
print("\n" + "="*40 + "\n")

# 2. Using .iloc (Integer position-based selection)
# Selecting the exact same data, but using the numerical index (0-based) for rows and columns.
# 'INV-01' is at index 0, 'INV-03' is at index 2. 'Product' is at index 0, 'Price' is at index 1.
iloc_selection = df.iloc[[0, 2], [0, 1]]

print("=== Selection using .iloc ===")
print("Query: df.iloc[[0, 2], [0, 1]]")
print("-" * 30)
print(iloc_selection)
print("\n")
```
