```python
import json
import os

# --- Setup ---
# Creating a dummy JSON file for the script to read.
# In a real environment, you would just read an existing file.
dummy_json_content = {
    "name": "Mantra Jhalani",
    "role": "Software Engineer",
    "project": "Krishi mitra",
    "skills": ["python", "c++", "flutter", "figma"]
}
with open("data.json", "w") as f:
    json.dump(dummy_json_content, f)


# Task 2: Extract a Small Report from a JSON File


def extract_json_report(filepath: str):
    """Reads a JSON file, transforms data, and prints a formatted report."""
    
    # 1. Load the simple JSON file
    with open(filepath, "r") as file:
        data = json.load(file)
    
    # 2. Use a list comprehension for a compact transformation
    # (Capitalizing the first letter of each skill for better presentation)
    display_skills = [skill.capitalize() for skill in data.get("skills", [])]
    
    # 3. Print selected values using f-strings
    print(f"\n===== SMALL REPORT =====")
    print(f" Name    : {data.get('name')}")
    print(f" Role    : {data.get('role')}")
    print(f" Project : {data.get('project')}")
    print(f" Skills  : {', '.join(display_skills)}")
    print(f"========================\n")

if __name__ == "__main__":
    extract_json_report("data.json")
    
    # --- Cleanup ---
    # Removing the temporary dummy file after execution
    if os.path.exists("data.json"):
        os.remove("data.json")
```
