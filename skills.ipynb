```python
# skills_counter.py - Lists skills, numbers them, and prints the total count.

# --- 1. Create the list ---
skills = ["Python" , "C++" , "flutter" , "HTML" , "CSS"]
print("my current skills:")
print("-" * 18)

# --- 2. Loop through using enumerate ---
# enumerate hands us the 'number' and the 'skill_name' for each item in the list
for indx , skills_name in enumerate(skills , start = 1):
    print(f"{indx}. {skills_name}")

print("-" * 18)

#--- 3. Print the total count using len() ---
total_skills = len(skills)
print(f"Total skill counted: {total_skills}")

 
```
