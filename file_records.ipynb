```python
import csv

def main():
    # ==========================================
    # TASK 1: Create students.csv
    # ==========================================
    
    # Sample data to populate the file
    headers = ['name', 'math', 'science', 'english']
    students_data = [
        ['Alice', 92, 88, 95],
        ['Bob', 75, 78, 72],
        ['Charlie', 50, 60, 55],
        ['Diana', 85, 90, 89]
    ]

    # Open file in write mode. newline='' prevents extra blank lines in Windows.
    with open('students.csv', mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(headers)       # Write the header row
        writer.writerows(students_data) # Write all student data rows
    
    print("Successfully created 'students.csv'.")

    # ==========================================
    # TASK 2: Read it back and calculate averages
    # ==========================================
    
    processed_results = [] # List to hold dictionaries for our DictWriter later

    with open('students.csv', mode='r') as file:
        reader = csv.reader(file)
        next(reader) # Skip the header row
        
        for row in reader:
            name = row[0]
            # Convert string representations of numbers from the CSV into integers
            math = int(row[1])
            science = int(row[2])
            english = int(row[3])
            
            # Calculate average
            average = (math + science + english) / 3
            
            # Determine Grade
            if average >= 90:
                grade = 'A'
            elif average >= 80:
                grade = 'B'
            elif average >= 70:
                grade = 'C'
            elif average >= 60:
                grade = 'D'
            else:
                grade = 'F'
                
            # Append as a dictionary for the DictWriter
            processed_results.append({
                'name': name,
                'average': round(average, 2), # Round to 2 decimal places for neatness
                'grade': grade
            })
            
    print("Successfully calculated averages and grades.")

    # ==========================================
    # TASK 3 & EXPLORE: Write results.csv using DictWriter
    # ==========================================
    
    result_headers = ['name', 'average', 'grade']
    
    with open('results.csv', mode='w', newline='') as file:
        # Initialize DictWriter with the file object and the expected keys (fieldnames)
        dict_writer = csv.DictWriter(file, fieldnames=result_headers)
        
        dict_writer.writeheader() # Automatically writes ['name', 'average', 'grade']
        
        # Write the rows using our list of dictionaries
        for result in processed_results:
            dict_writer.writerow(result)
            
    print("Successfully generated 'results.csv' using csv.DictWriter.")

# Run the script
if __name__ == "__main__":
    main()
```
