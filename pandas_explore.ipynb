```python
import pandas as pd


# TASK: Create the DataFrame

data = {
    'name': ['Amit', 'Priya', 'Rahul', 'Sneha', 'Vikram', 'Anjali', 'Karan', 'Neha', 'Rohan', 'Pooja'],
    'city': ['Jaipur', 'Jodhpur', 'Jaipur', 'Udaipur', 'Kota', 'Jaipur', 'Jodhpur', 'Bikaner', 'Udaipur', 'Jaipur'],
    'math_score': [85, 92, 70, 88, 76, 95, 60, 82, 78, 91],
    'science_score': [80, 89, 75, 91, 85, 96, 65, 88, 80, 94],
    'english_score': [78, 85, 82, 89, 80, 92, 70, 86, 75, 88]
}

df = pd.DataFrame(data)

# Pre-computation: Add a total_score column to make querying cleaner
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score']

print("--- Initial Dataset (First 5 Rows) ---")
print(df.head(), "\n")

#Task

print("1. Average score in each subject:")
print(df[['math_score', 'science_score', 'english_score']].mean(), "\n")

print("2. Student with the highest total score:")
print(df.loc[df['total_score'].idxmax()], "\n")

print("3. Number of students from each city:")
print(df['city'].value_counts(), "\n")

print("4. Students with a math score above 75:")
# Selecting just a few columns to make the printed output cleaner to read
print(df[df['math_score'] > 75][['name', 'city', 'math_score']], "\n")


# EXPLORE: df.sort_values() vs df.nlargest()
# ==========================================
# df.nlargest() is generally more performant than sorting the entire dataframe 
# just to get the top few rows.

print("--- EXPLORE: Top 3 Students by Total Score ---")
print(df.nlargest(3, 'total_score')[['name', 'city', 'total_score']])
```
