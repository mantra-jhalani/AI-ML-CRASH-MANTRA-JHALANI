```python
class Student:
    # 1. Class Variable: Shared by all instances of the Student class
    school_name = "Tech Academy"

    # 2. __init__ method to set up instance variables
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks  # Expecting a list of at least 3 numbers

    # 3. Method to calculate the average of the marks
    def average(self):
        if not self.marks:  # Safety check in case the list is empty
            return 0
        return sum(self.marks) / len(self.marks)

    # 4. Method to determine the grade based on the average
    def grade(self):
        avg = self.average()
        
        # Standard grading scale
        if avg >= 90:
            return 'A'
        elif avg >= 80:
            return 'B'
        elif avg >= 70:
            return 'C'
        elif avg >= 60:
            return 'D'
        else:
            return 'F'

    # 5. __str__ method to format how the object is printed
    def __str__(self):
        # We access the class variable via Student.school_name 
        # (though self.school_name also works)
        return (f"[{Student.school_name}] Roll: {self.roll_no} | "
                f"Name: {self.name.ljust(15)} | "
                f"Avg: {self.average():.1f} | Grade: {self.grade()}")


# 6. Driver Code: Creating and Printing Objects


# Creating 3 Student objects with at least 3 subjects each
student1 = Student("Alice Smith", 101, [95, 88, 92])
student2 = Student("Bob Johnson", 102, [72, 65, 78])
student3 = Student("Charlie Brown", 103, [55, 60, 58])

# Printing each object. This automatically calls the __str__ method.
print(student1)
print(student2)
print(student3)
```
